```python
#Functions 
#sg ({0, 1}, optional) – Training algorithm: 1 for skip-gram; otherwise CBOW.
def wordEmbedding(corpus,modelType,size):
    all_sentences = nltk.sent_tokenize(corpus)
    all_words = [nltk.word_tokenize(sent) for sent in all_sentences]
    model = Word2Vec(all_words, size=size, window=5, min_count=5, workers=4,sg=modelType)
    return model 

def tokenization(filename):
    tokens=nltk.word_tokenize(filename)  
    return tokens    

def wordCount(wList):
    wordFreq = []
    for w in wList :
        wordFreq.append(wList.count(w))
    return wordFreq 

#Find most frequet words 
def mostrFreq(corpus):
    sinhala = tokenization(corpus)
    sinWrdFreq=wordCount(sinhala)
    dfSin=pd.DataFrame({"word":sinhala,"freq":sinWrdFreq}).sort_values(by='freq',ascending=False)
    dfSin=dfSin[~(dfSin['word']=='')]
    dfSin=dfSin.drop_duplicates()
    dfSin=dfSin[~((dfSin['word'].str.len())==1)]
    return dfSin

def similarityTrained(sinHighFreqWords,model):
    for x in range(len(sinHighFreqWords)):
        y=sinHighFreqWords[x]
        print(model.wv.similar_by_vector(model.wv[y], topn=10, restrict_vocab=None))
        print("\n")

def similarityTxt(sinHighFreqWords,model):
    for x in range(len(sinHighFreqWords)):
        y=sinHighFreqWords[x]
        print(model.wv.similar_by_vector(model.wv[y], topn=10, restrict_vocab=None))
        print("\n")
        
def similarityBin(sinHighFreqWords,model):
    for x in range(len(sinHighFreqWords)):
        y=sinHighFreqWords[x]
        print(model.most_similar(positive=y, topn=10, restrict_vocab=None))
        print("\n")
```

```python
sinHighFreqWords=['මුදල්','නිවාඩු','භාෂා','වැඩසටහන්','ආයතන','වාර්තාව','වියදම්','කාර්යාලය','කාලය','දෙපාර්තමේන්තුව']
```

```python
#1) Fiding top10 similar word to given common words in sinHighFreqWords(Skipgram model)

#Skip-gram model trainning 
skipGram=wordEmbedding(corpus1,1,100)

#Fiding similarity
similarityTrained(sinHighFreqWords,skipGram)
```

```python
#2) Fiding top10 similar word to given common words in sinHighFreqWords(CBOW model)

#CBOW model trainning 
cbow=wordEmbedding(corpus1,0,100)

#Fiding similarity
similarityTrained(sinHighFreqWords,cbow)
```

```python
#3) Pre-trained model (Txt version)
filename = 'cc.si.300.vec'
modelTxt = KeyedVectors.load_word2vec_format(filename, binary=False)

#Fiding similarity
similarityTxt(sinHighFreqWords,modelTxt)
```
